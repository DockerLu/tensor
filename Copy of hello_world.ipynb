# Apply to Actuarial Science (Solvency prediction)

```
import pandas as pd
import tensorflow as tf
import numpy as np
```

```
df=pd.read_csv('alldata _t-2 t complete_.csv')
```

```
df
```

```
x = df.iloc[:,1:16]
```

```
y = df['Y0']
```

```
msk = np.random.rand(len(df)) < 0.8
```

```
x_train = x[msk]
x_test = x[~msk]
print(len(x_train))
print(len(x_test))
print(len(x))
```

```
y_train = y[msk]
y_test = y[~msk]
print(len(y_train))
print(len(y_test))
print(len(y))
```

```
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Flatten(input_shape=(15,)))
model.add(tf.keras.layers.Dense(10,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(5,activation=tf.nn.relu))
#model.add(tf.keras.layers.Dense(5,activation=tf.nn.relu))
#model.add(tf.keras.layers.Dense(5,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(2,activation=tf.nn.softmax))

model.compile(optimizer ='adam',
            loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

early_stopping_monitor = tf.keras.callbacks.EarlyStopping(patience = 2)
model.fit(x_train, y_train, epochs=3)
```

```
model.save('trained_robot.model')
```

```
new_model = tf.keras.models.load_model('trained_robot.model')
```

```
predictions = new_model.predict(x_test)
```

```
predictions
```

```
y_test_list=y_test.tolist()
y_test_list
```

```
count = 0
for i in range(len(y_test_list)):
  t = np.argmax(predictions[i])
  if t == y_test_list[i]:
    count =count+1

prob =count/float(len(y_test_list))
print(prob)
```
